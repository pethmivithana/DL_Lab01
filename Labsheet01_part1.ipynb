```python
#1)
import numpy as np

array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print(array_4x4)
print()
#2)
import matplotlib.pyplot as plt

exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.normal(size=100000)

plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.axis([-2.5, 10, 0, 1.5])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
print()

#3)
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Z = X² + Y²')
plt.show()
print()

#4)
# Pandas for managing datasets
import pandas as pd

# Matplotlib for plotting
from matplotlib import pyplot as plt
%matplotlib inline

# Seaborn for plotting
import seaborn as sns

# Download dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

# Load dataset into DataFrame 'df'
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Display first 5 rows
df.head()

features = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Pearson Correlation
pearson_corr = features.corr(method='pearson')
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation Heatmap")
plt.show()

# Spearman Correlation
spearman_corr = features.corr(method='spearman')
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm')
plt.title("Spearman Correlation Heatmap")
plt.show()

```
